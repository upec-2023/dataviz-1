### Scatter plot (nuage de points) 

>La fonction **``scatter``** permet de tracer des nuages de points. Elle s'utilise d'une manière similaire à ``plot``.<br>
>
> <span style="color:#09b038; text-decoration : underline"> Exemple : Utilisation de la méthode **scatter**</span><br> 

> ```python
>plt.scatter(range(0,7), [8,7,6,5,6,7,8], color='red', marker= '*', s=40)
>```
>
> produit le graphique :
<img src="https://datascientest.fr/train/assets/barp4.png" style="height:250px">


> Il est possible également de tracer deux nuages de points simultanément sur le même graphique en faisant deux fois appel à la commande `plt.scatter`  l'une après l'autre.<br>
Les paramètres ``c`` (pour la couleur) et ``s`` (pour la taille) peuvent être :
- **Soit** des éléments uniques (c = 'red') 
- **Soit** des listes de la même taille que le nombre de points (`c = ['red', 'green', 'blue']` si 3 points par exemple).

> La liste de couleurs disponibles sur Matplotlib pourrait sembler restreinte pour certains, il est possible d'utiliser des  codes couleurs [HEX ](https://www.color-hex.com/) pour colorer tout type de graphique.

* Créer la listes **``x``** des abscisses de ``1`` à ``5``, puis les listes **``y1``** et **``y2``** suivantes: ``[0,2, 3, 8, 15], [5, 8, 11, 16, 22]``.
* Représenter les points de coordonnées **``x*``** et **``y1``** en jaune et de tailles croissantes: ``30, 60, 90, 120, 150``.
* Représenter les points de coordonnées **``x``** et **``y2``**, de taille ``40`` et de couleurs: [#d248a2', '#8b48d2', '#48a5d2', '#4dd248', '#d1d248'].



````python
### Insérez votre code ici

x = np.arange(1,6,1)

y1 = [0,2,3,8,15]
y2 = [5,8,11,16,22]

plt.scatter(x,y1,c='y',s=[30,60,90,120,150])
plt.scatter(x,y2,c=['#d248a2', '#8b48d2', '#48a5d2', '#4dd248', '#d1d248'],s=40);
````

> Lorsque l'on travaille avec des dataframes, le paramètre **`c`** permet de colorer les points par rapport aux valeurs prises par les points dans une colonne, en insérant cette colonne en argument au paramètre **`c`**. 

* Lire le fichier `'sales_data.csv'`.
* Afficher un nuage de points d'abscisses **``df.Product1``** et d'ordonnées **``df.Product2``**, de taille 30, et en colorant les points par rapport à **`df.Returns`**.


````python
### Insérez votre code ici
df = pd.read_csv('sales_data.csv')
print(df.head())


plt.scatter(df['Product1'],df['Product2'],s=30,c=df['Returns'])
plt.xlabel('Product1')
plt.ylabel('Product2');
````

<div class="alert alert-success">
<i class="fa fa-question-circle" aria-hidden="true"></i>&emsp; 
Voici le graphique équivalent obtenu directement depuis la méthode <b><code style = "background-color: transparent ; color : inherit">plot.scatter()</code></b> des <i>dataframes</i> pandas.

````python
df.plot.scatter(x='Product1', y='Product2', c = 'Returns', s=30, cmap = 'viridis');

# L'ajout de cmap permet de choisir une palette de couleurs, parmi toutes celles disponibles.
````
