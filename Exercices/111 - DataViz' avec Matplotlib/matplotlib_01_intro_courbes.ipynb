### Introduction

>Bienvenue dans ce cours dédié aux bases de la *Data Visualization* avec Python grâce à sa principale bibliothèque dédiée : **Matplotlib**. <br>
Matplotlib est un puissant outil pour tracer des graphiques et visualiser des données.
Elle est destinée à produire des graphiques de toutes sortes (voir [ici](http://matplotlib.org/gallery.html) pour une galerie d'images produites avec Matplotlib)

> **``Matplotlib``** contient la sous-bibliothèque **``pyplot``** qui crée une interface proche de celle du logiciel commercial **Matlab** et qui contient des fonctions très similaires à celui-ci.


> Par ailleurs, Matplotlib est une bibliothèque très riche, toutes ses fonctionnalités ne peuvent être traitées. Cette formation fait le choix d'explorer certaines fonctions plus que d'autres, l'objectif général étant de rendre un datascientist autonome sur le module en fin de formation.

>Une fois un graphique construit, c'est l'instruction **``plt.show()``** qui permettra de le visualiser.<br>
