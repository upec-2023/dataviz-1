### Boxplot (boîte à moustache) 

>Les boîtes à moustaches (ou boxplots) sont des graphiques très appréciés et utilisés, notamment lors d'analyses descriptives de données continues.<br>
> La fonction **``plt.boxplot``** permet d'afficher des boites à moustaches sur un graphique.<br>
Elle prend en argument une ou plusieurs séquences de valeurs réunies dans une autre séquence.<br>
La boîte principale donne les limites du premier et dernier quartile, la barre rouge du milieu indique la médiane.<br> Les moustaches vont jusqu'à la valeur la plus extrême dans la limite de 1.5 fois la hauteur de la boîte. Les points au-delà des moustaches sont représentés par des +, ce sont des *outliers*.

<span style="color:#09b038; text-decoration : underline"> Exemple :</span><br>
```python
    plt.boxplot([[1, 2, 3, 4, 5], [7, 5, 8, 4, 9, 5, 7]])
```

### <i class="fas fa-chart-pie"></i> Camembert (Pie Charts) 

> Les camemberts sont des diagrammes circulaires divisés par secteurs (*wedges*). C'est une manière efficace de représenter l'information lorsque l'on s'intéresse principalement à la comparaison d'un secteur avec le camembert tout entier plutôt qu'entre deux secteurs.<br>

>La fonction **``pie``** permet de tracer des diagrammes circulaires à partir d'une séquence **``X``**.  
Chaque élément **$x_{i}$** de **``X``** est représenté par une part proportionnelle à **$x_{i}$/sum(X**).
Si la somme des éléments de **``X``** est inférieure à 1, le camembert est représenté à l'aide des valeurs de **``X``** directement, aucune normalisation n'est effectuée, ce qui laisse un espace vide.

<div class="alert alert-success">
<i class="fa fa-question-circle"></i> &emsp; 
Les  diagrammes circulaires s'affichent mieux si la figure présente un format agrandi et <b>carré</b>.<br>
Lorsque l'on utilise une fonction de <b>pyplot</b>, une figure se crée automatiquement, au format standard par défaut.<br>
La figure peut être crée aux dimensions <b>(h,l)</b> grâce à la commande <b>plt.figure( figsize= (h, l) )</b>.
</div>

> De nombreux paramètre permettent de customiser son camembert.<br>
* ``explode``: liste de la même taille que la séquence de données, permet d'éloigner une ou plusieurs part du centre en indiquant de quelle fraction de rayon chaque part doit être éloignée (0 par défaut). 
* ``colors``: séquence de couleurs à utiliser pour les parts.
* ``labeldistance`` : la distance des labels au centre (> 1 pour être à l'extérieur du cercle).
* ``autopct`` : une fonction (**lambda**) qui prend le pourcentage calculé pour une part et renvoie ce qui doit être affiché pour ce pourcentage.
* ``pctdistance`` : la distance au centre à laquelle le pourcentage précédent doit être affiché (1 = sur le cercle).
* ``shadow = True`` : indique qu'il faut afficher une ombre.
