> Pour ajouter une annotation avec une flèche descriptive pointant vers un point précis du graphique, on utilise `plt.annotate`.<br>
Cette méthode prend les arguments suivants : 
>>- le texte que l'on veut afficher.<br>
>>- `xy`, qui indique les coordonnées où se trouve le point à annoter.<br>
>>- `xytext`, qui indique les coordonnées du point où démarre le texte.<br>
>>- ``arrowprops``, qui sont les propriétés de la flèche d'annotation entre { } : couleurs, taille de la flèches, style de flèche, etc...

> <span style="color:#09b038; text-decoration : underline"> Exemple : Utilisation de la commande **plt.annotate**</span><br> 


```python
plt.annotate('Limite', xy=(1, 2), xytext=(1, 2.5), arrowprops={'facecolor':'blue'} )
```
