> Pour superposer deux graphiques au sein d'une même figure, il est possible d'utiliser l'attribut `ax` d'un objet retourné par seaborn, afin de l'indiquer comme axe au nouveau graphique crée.<br>

```py
p = sns.catplot(x= "var1", y=var2,  data=df)
sns.swarmplot(x= "var1", "var2", data= df, ax= p.ax) #l'argumnt ax= p.ax permet de superposer les graphiques
```
